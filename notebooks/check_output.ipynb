```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('../data/results.csv')
df.sample(2)
```

```python
lens = []
years = []
for county in df.County.unique():
    temp = df[df.County == county]
    for y in temp.Year.unique():
        years.append(y)
    lens.append(temp.shape[0])

print(Counter(lens))
print(Counter(years))
```
